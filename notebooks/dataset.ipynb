```python
headers = pd.read_csv('../data/spambase_headers.csv', header = 0, index_col=False).squeeze();
headers = headers.columns.tolist()
df.columns = headers
df.head(10)
```

```python
df["is_spam"].value_counts(normalize=True)
```

```python
for col in df.columns[:-1]:
        df[col] = df[col] / df[col].max()

        # Rename columns to remove prefix/suffix
        new_col_name = re.sub('^word_freq_|^char_freq_|^capital_run_length_', '', col)
        df = df.rename(columns={col: new_col_name})

df.head()
```
